```python
# I)
def rec1(n):
    if (n <= 1):
        return 1
    else:  
        return rec1 (n-1) + rec1 (n-1) 

# Su notación Big-Oh es O(n**2) por su parte recursiva, en la que se llama 2 veces

import time

# Función para medir el tiempo de ejecución
def medir_tiempo(func, *args):
    inicio = time.time()
    resultado = func(*args)
    fin = time.time()
    tiempo_ejecucion = fin - inicio
    print("Tiempo de ejecución:", tiempo_ejecucion)
    return resultado

# Ejemplo de uso
n = 30
resultado = medir_tiempo(rec1, n)
print("Resultado:", resultado)

"""
Tiempo de ejecución: 126.55212211608887
Resultado: 536870912
"""
```

```python
# III)
def f(n):
    if (n==0):
        return 0
    elif ( n == 1 ):
        return 1
    else: 
        return f(n-2) * f (n-2)
    
#Su notación Big O es O(n/2) o simplemente O(n)
```

```python
# IV)
def potencia_iter (x, n):
    if (n==0):
        potencia = 1
    elif (n==1):
        potencia = x
    else:
        potencia = x
        for i in range(2,n):
            potencia = potencia * x
    return potencia

#Su notación Big O es O(n) (lineal) por el ciclo for
```

```python
# V) 
def potencia_rec(x, n):
    if (n==0):
        return 1
    elif (n==1):
        return x
    elif (n % 2) == 0:
        return potencia_rec(x * x, n/2)
    else:
        return potencia_rec(x * x, n/2) * x

#Su notación Big O es O(log n) ya que usa la estrategia divide y venceras
```

Dada la recurrencia:

T(n) = 
  2                      si n = 1,
  T(n-1) + n    si n ≥ 2.

Comencemos resolviendo los primeros términos de la recurrencia:

T(1) = 2
T(2) = T(1) + 2 = 2 + 2 = 4
T(3) = T(2) + 3 = 4 + 3 = 7
T(4) = T(3) + 4 = 7 + 4 = 11

Podemos notar un patrón emergente en los valores de T(n). Parece que T(n) está relacionado con la suma de los números naturales desde 1 hasta n.

De hecho, podemos demostrarlo mediante el método de inducción matemática:

1. Caso base:
   Verificamos si la fórmula es válida para el caso base, que es T(1):
   T(1) = 2

2. Hipótesis de inducción:
   Suponemos que la fórmula es válida para un número arbitrario k, es decir, T(k) = 1 + 2 + ... + k.

3. Paso de inducción:
   Demostramos que la fórmula también se cumple para el número k+1, es decir, T(k+1) = 1 + 2 + ... + (k+1).

   Utilizando la hipótesis de inducción, tenemos:
   T(k+1) = T(k) + (k+1)
          = 1 + 2 + ... + k + (k+1)
          = 1 + 2 + ... + (k+1)

Por lo tanto, la fórmula se cumple para k+1.

Basándonos en esta observación y utilizando la fórmula de la suma de una serie aritmética, podemos concluir que la solución de la recurrencia es:

T(n) = 1 + 2 + ... + n = (n(n+1))/2

Por lo tanto, la fórmula T(n) = (n(n+1))/2 es la solución de la recurrencia dada.

Para llegar a la fórmula T(n) = (n(n+1))/2 solo analizando los ejemplos donde n va de 1 al 6, podemos observar los valores correspondientes y buscar un patrón.

Los primeros números triangulares son:

T(1) = 1
T(2) = 1 + 2 = 3
T(3) = 1 + 2 + 3 = 6
T(4) = 1 + 2 + 3 + 4 = 10
T(5) = 1 + 2 + 3 + 4 + 5 = 15
T(6) = 1 + 2 + 3 + 4 + 5 + 6 = 21

Analizando estos ejemplos, podemos notar un patrón: el número triangular T(n) es igual a la suma de todos los números naturales desde 1 hasta n.

Si expresamos la suma de los números naturales desde 1 hasta n de manera algebraica, obtenemos la fórmula:

T(n) = 1 + 2 + 3 + ... + n

Esta suma puede ser reescrita utilizando la propiedad de la suma de una progresión aritmética. En una progresión aritmética, la suma de los n primeros términos se calcula mediante la fórmula (n(a1 + an))/2, donde a1 es el primer término y an es el último término.

En nuestro caso, el primer término a1 es 1 y el último término an es n. Por lo tanto, la suma de los n primeros términos es:

T(n) = (n(1 + n))/2

Simplificando esta expresión, llegamos a la fórmula final:

T(n) = (n(n+1))/2

Esta fórmula nos permite calcular de manera directa el número triangular correspondiente a cualquier valor de n.

Así, al analizar los ejemplos donde n va de 1 al 6, podemos identificar el patrón y derivar la fórmula general para los números triangulares.
