### Ejercicio 1: Búsqueda Ternaria

link: https://es.wikipedia.org/wiki/B%C3%BAsqueda_ternaria

La **búsqueda ternaria** determina que el extremo que se busca, no puede estar en el primer tercio del dominio o que no puede estar en el último tercio del dominio, luego se repite el proceso en los dos tercios restantes. Es decir, dividiremos la lista en 3 *'sub-listas'*. 

    a) Implementar el Algoritmo de Búsqueda Ternaria.


```python
"""
La búsqueda ternaria es un algoritmo de búsqueda utilizado en informática para buscar un valor en un conjunto de datos ordenados. Es similar a la búsqueda binaria, pero en lugar de dividir el conjunto en dos partes iguales, lo divide en tres partes.

El funcionamiento básico de la búsqueda ternaria es el siguiente:

El conjunto de datos debe estar ordenado de forma ascendente o descendente.
Se toman dos puntos, generalmente llamados "izquierda" y "derecha", que delimitan el rango de búsqueda inicial.
Se calcula el punto intermedio entre los puntos izquierdo y derecho, dividiendo el rango en tres partes aproximadamente iguales.
Se compara el valor buscado con el valor en el punto intermedio.
Si el valor buscado es igual al valor en el punto intermedio, se ha encontrado el elemento y se devuelve su posición.
Si el valor buscado es menor que el valor en el punto intermedio, la búsqueda se realiza en la primera parte del rango (entre los puntos izquierdo y medio).
Si el valor buscado es mayor que el valor en el punto intermedio, la búsqueda se realiza en la tercera parte del rango (entre los puntos medio y derecho).
Se repiten los pasos 3 a 7 hasta que se encuentre el valor buscado o se reduzca el rango de búsqueda a cero.
"""

# Algoritmo de búsqueda ternario para devolver la posición de
# objetivo `x` en una lista dada `A`, ordenada de forma ascendente

def busqueda_ternaria(lista, x):
    izq, der = 0, len(lista) - 1

    while izq <= der:
        tercio = (der - izq) // 3
        medio1 = izq + tercio
        medio2 = der - tercio

        if lista[medio1] == x:
            return medio1
        elif lista[medio2] == x:
            return medio2
        elif x < lista[medio1]:
            der = medio1 - 1
        elif x > lista[medio2]:
            izq = medio2 + 1
        else:
            izq = medio1 + 1
            der = medio2 - 1

    return -1

def busqueda_ternaria_lineal(lista,x):
    izq, der = 0, len(lista) - 1
    tercio = (der - izq) // 3
    medio1 = izq + tercio
    medio2 = der - tercio
    if lista[medio1] == x:
        return medio1
    elif lista[medio2] == x:
        return medio2
    elif x<lista[medio1]:
        i=izq
        for z in lista[izq:medio1]:
            if x==z:
                return i
            i+=1
        return -1
    elif x > lista[medio1] and x < lista[medio2]:
        i=medio1
        for z in lista[medio1:medio2]:
            if x==z:
                return i
            i+=1
        return -1
    elif x>medio2:
        i=medio2
        for z in lista[medio2:der+1]:
            if x==z:
                return i
            i+=1
        return -1
    
def busqueda_ternaria_recursiva(lista, x):
    if len(lista) == 0:
        return -1

    tercio = len(lista) // 3
    medio1 = tercio
    medio2 = 2 * tercio

    if lista[medio1] == x:
        return medio1
    elif lista[medio2] == x:
        return medio2
    elif x < lista[medio1]:
        return busqueda_ternaria_recursiva(lista[:medio1], x)
    elif x > lista[medio2]:
        resultado = busqueda_ternaria_recursiva(lista[medio2 + 1:], x)
        if resultado == -1:
            return -1
        return medio2 + 1 + resultado
    else:
        resultado = busqueda_ternaria_recursiva(lista[medio1 + 1 : medio2], x)
        if resultado == -1:
            return -1
        return medio1 + 1 + resultado


L=(0,1,2,3,4,5,6,7,8)
print(busqueda_ternaria(L,6))
print(busqueda_ternaria_lineal(L,6))
print(busqueda_ternaria_recursiva(L,6))
```

### Ejercicio 2: Buble Sort 

link: https://sites.google.com/site/programacioniiuno/temario/unidad-6---anlisis-de-algoritmos/algoritmo-bubblesort
El algoritmo de ordenamiento por burbujeo (o bubblesort) es un algoritmo "simple" de ordenamiento. La idea es que se van a comparar pares consecutivos de elementos, y en caso en que estén desordenados, es decir que el primer elemento (izquierda) sea mayor al segundo (derecha), se hará un intercambio de posiciones.

Entonces, si tenemos que comparar cada elemento con el siguiente, podríamos pensar que el algoritmo es:

Iterar cada elemento de la lista (hasta el anteúltimo), e ir comparándolo con el siguiente.
Cambiarlos de posiciones si están desordenados

Entonces, no nos alcanza con una única iteración. Así que hacemos otra implementación del algoritmo que haga lo mismo que la anterior, solo que varias veces. Cuántas ? N, donde N es la cantidad de elementos que tiene la lista.

<img src="https://drive.google.com/uc?id=1INViHe9uLtR3atFoLKl_D6UCzyX4ex4G" align="right" />

Algoritmo de ordenamiento. Burbuja(Bubble Sort en inglés) es un sencillo algoritmo de ordenamiento. Funciona revisando cada elemento de la lista que va a ser ordenada con el siguiente, intercambiándolos de posición si están en el orden equivocado. Es necesario revisar varias veces toda la lista hasta que no se necesiten más intercambios, lo cual significa que la lista está ordenada. Este algoritmo obtiene su nombre de la forma con la que suben por la lista los elementos durante los intercambios, como si fueran pequeñas burbujas. También es conocido como el método del intercambio directo. Dado que solo usa comparaciones para operar elementos, se lo considera un algoritmo de comparación, siendo el más sencillo de implementar.

    a) Implementar el Algoritmo de Ordenamiento Burbuja. 

#### Ejemplo:

<img src="https://drive.google.com/uc?id=1HFKRTLd6UiSAe7b1pPtekQLxg5xk5dDL" />


```python
def bubble_Sort(lista):
    for _ in range(len(lista)):
        for i in range(len(lista) - 1):
            if lista[i] > lista[i+1]:
                lista[i], lista[i+1] = lista[i+1], lista[i]
    
    return lista

L=[4,3,7,1,2,6,9,8,5]
bubble_Sort(L)
```

### Ejercicio 3:  Ordenamiento Shell (Shell Sort)

link: https://es.wikipedia.org/wiki/Ordenamiento_Shell

El Shell sort es una generalización del ordenamiento por inserción, teniendo en cuenta dos observaciones:

1) El ordenamiento por inserción es eficiente si la entrada está "casi ordenada".

2) El ordenamiento por inserción es ineficiente, en general, porque mueve los valores solo una posición cada vez.


El algoritmo Shell sort mejora el ordenamiento por inserción comparando elementos separados por un espacio de varias posiciones. Esto permite que un elemento haga "pasos más grandes" hacia su posición esperada. Los pasos múltiples sobre los datos se hacen con tamaños de espacio cada vez más pequeños. El último paso del Shell sort es un simple ordenamiento por inserción, pero para entonces, ya está garantizado que los datos del vector están casi ordenados.


    a) Implementar el Algoritmo de Ordenamiento Shell. 

```python
def shell_sort(lista):
    n=len(lista)//2 # Seleccionar una brecha inicial
    
    while n>0:
        for i in range(n,len(lista)): # Iterar sobre los elementos desde la brecha hasta el final del arreglo
            
            while i>=n and lista[i-n]>lista[i]: # Realizar comparaciones y desplazamientos en el subarreglo
                lista[i], lista[i - n] = lista[i - n], lista[i] # Intercambiar los elementos
                i-=n # Mover hacia atrás en el subarreglo
    
        n //=2  # Reducir la brecha a la mitad en cada iteración
    return lista

L=[4,3,7,1,2,6,9,8,5]
shell_sort(L)      
```

### Ejercicio 4: Quick Sort Mejorado,


Implemnetar el algoritmo de ordenamiento Quick Sort, utlizando el método de *'mediana de tres'* para la selección del pivote. 



```python
def quick_sort(lista):
    if len(lista) < 2: 
        return lista

    menores, medio, mayores = _partition(lista)
    return quick_sort(menores) + medio + quick_sort(mayores)

def mediana_de_tres(lista):
    a=lista[0]
    b=lista[len(lista)-1]
    c=lista[len(lista)//2]
    if a <= b <= c or c <= b <= a:
        return b
    elif b <= a <= c or c <= a <= b:
        return a
    else:
        return c

def _partition(lista):
    pivote = mediana_de_tres(lista)
    menores, medio, mayores = [], [], [] 
    
    for x in lista:
        if x < pivote: 
            menores.append(x)
        elif x > pivote: 
            mayores.append(x)
        else:
            medio.append(x)
      
    return menores, medio, mayores

L=[4,3,7,1,2,6,9,8,5]
quick_sort(L)
```
