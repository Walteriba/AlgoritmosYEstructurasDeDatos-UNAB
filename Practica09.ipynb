Implementar una Pila utilizando la siguiente definición de una *Lista 'Simple' Enlazada*. Utilizar el prototipo de la clase Lista Simple Enlazada mostrado a continuación. 

```python
class PilaEnlazada:
    """Clase Lista 'Simplemente' Enlazada."""

#-------- Clase Anidada - NODO ------------#
    class _Node:
        """Clase Nodo."""
        __slots__ = '_element', '_next' # optimiza el uso de memoria

        def __init__(self, element, next): 
            self._element = element # inicializar el contenido del Nodo
            self._next = next       #referencia al siguiente Nodo

    def __init__(self):
        """Crea una Pila Vacia.""" 
        self._head = None  #referencia a la "cabeza" de la lista
        self._size = 0     #cantidad de elementos
    
    
    def __len__(self):
        """Retorna el numero de elementos de la Pila."""
        return self._size

    def is_empty(self):
        """Retorna VERDADERO si la Pila esta Vacia."""
        return self._size == 0

    #----------- Métodos de la Pila a completar ----------- #  
    
    def push(self, element):
        nuevo_nodo = self._Node(element, None) #creo un nodo nuevo, siguiendo la forma de la clase anidada
        nuevo_nodo._next = self._head          # sin embargo en esta forma, no recorro la lista con un while, sigo la forma
        self._size+=1                          # vista en la bibliografia
 
    def push(self, element): #forma chatgtp
        """Agrega un elemento al final de la lista (como en una pila)."""
        new_node = self._Node(element, None)  # Crea un nuevo nodo con el elemento y sin referencia al siguiente nodo
        if self.is_empty():
            self._head = new_node  # Si la lista está vacía, el nuevo nodo se convierte en la cabeza de la lista
        else:
            current = self._head
            while current._next is not None:
                current = current._next
            current._next = new_node  # Agrega el nuevo nodo al final de la lista
        self._size += 1  # Incrementa el tamaño de la lista
    
    def pop(self): #forma segun bibliografia
        x=self._head._element
        self._head = self._head._next
        self._size -= 1
        return x 
 
    def pop(self): #chatgtp
        """Elimina y retorna el elemento en el tope de la Pila."""
        if self.is_empty():
            raise IndexError("La Pila está vacía")
        current = self._head
        prev = None
        while current._next is not None:
            prev = current
            current = current._next
        if prev is None:
            self._head = None  # Si solo había un nodo en la lista
        else:
            prev._next = None  # Elimina la referencia al último nodo
        self._size -= 1  # Reduce el tamaño de la lista
        return current._element
    
    def top(self): #forma segun blibliografia
        if self._head is not None:
            return self._head._element
        else:
            return None
    
    def top(self): #chatgtp
        """Retorna el elemento en el tope de la Pila sin eliminarlo."""
        if self.is_empty():
            raise IndexError("La Pila está vacía")
        current = self._head
        while current._next is not None:
            current = current._next
        return current._element
```

### Ejercicio 5

Implementar el tipo de dato lista enlazada circular.

```python
class ListaEnlazadaCircular:
    """Clase Lista enlazada Circular."""

#-------- Clase Anidada - NODO ------------#
    class _Node:
        """Clase Nodo."""
        __slots__ = '_element', '_next' # optimiza el uso de memoria

        def __init__(self, element, next): 
            self._element = element # inicializar el contenido del Nodo
            self._next = next       # referencia al siguiente Nodo

    def __init__(self):
        """Crea una Lista Vacia.""" 
        self._head = None  #referencia a la "cabeza" de la lista
        self._size = 0     #cantidad de elementos
     
    
    def __len__(self):
        """Retorna el numero de elementos de la Lista."""
        return self._size

    def is_empty(self):
        """Retorna VERDADERO si la Lista esta Vacia."""
        return self._size == 0
    
    def append(self, x):
        """Agrega un elemento al final de la lista."""
        new_node = self._Node(x, None)  # Crear un nuevo nodo con el elemento x

        if self.is_empty():
            # Caso especial: la lista está vacía
            new_node._next = new_node  # El nuevo nodo apunta a sí mismo
            self._head = new_node  # El nuevo nodo se convierte en la cabeza
        else:
            # Caso general: la lista no está vacía
            new_node._next = self._head  # El nuevo nodo apunta a la cabeza
            last_node = self._head
            while last_node._next != self._head:
                last_node = last_node._next  # Encontrar el último nodo de la lista
            last_node._next = new_node  # El último nodo apunta al nuevo nodo

        self._size += 1  # Incrementar el tamaño de la lista
    
    
    def insert(self, i, x):
        """Agrega el elemento 'x' en la posición 'i'."""
        if i < 0 or i > len(self):
            raise IndexError("Posición inválida")

        new_node = self._Node(x, None)  # Crear un nuevo nodo con el elemento x

        if i == 0:
            if self.is_empty():
                # Caso especial: la lista está vacía
                new_node._next = new_node  # El nuevo nodo apunta a sí mismo
                self._head = new_node  # El nuevo nodo se convierte en la cabeza
            else:
                # Caso general: insertar al principio
                last_node = self._head
                while last_node._next != self._head:
                    last_node = last_node._next  # Encontrar el último nodo de la lista
                new_node._next = self._head  # El nuevo nodo apunta a la cabeza actual
                self._head = new_node  # El nuevo nodo se convierte en la nueva cabeza
                last_node._next = self._head  # El último nodo apunta a la nueva cabeza
        else:
            prev_node = self._head
            for _ in range(i - 1):
                prev_node = prev_node._next  # Encontrar el nodo anterior a la posición i
            new_node._next = prev_node._next  # El nuevo nodo apunta al nodo siguiente
            prev_node._next = new_node  # El nodo anterior apunta al nuevo nodo

        self._size += 1  # Incrementar el tamaño de la lista
        
    def remove(self, x):
        """Eliminar la primera aparición de 'x' en la lista."""
        if self.is_empty():
            raise ValueError("La lista está vacía")

        if self._head._element == x:
            # Caso especial: el elemento a eliminar está en la cabeza
            if self._head._next == self._head:
                # Caso especial: la lista solo tiene un elemento
                self._head = None  # La lista queda vacía
            else:
                last_node = self._head
                while last_node._next != self._head:
                    last_node = last_node._next  # Encontrar el último nodo de la lista
                last_node._next = self._head._next  # El último nodo apunta al siguiente de la cabeza
                self._head = self._head._next  # El siguiente de la cabeza se convierte en la nueva cabeza
        else:
            prev_node = self._head
            curr_node = self._head._next
            while curr_node != self._head:
                if curr_node._element == x:
                    prev_node._next = curr_node._next  # Saltar el nodo actual
                    self._size -= 1  # Decrementar el tamaño de la lista
                    return
                prev_node = curr_node
                curr_node = curr_node._next

        raise ValueError("El elemento no está en la lista")

    def pop(self, i=None):
        """Borra el elemento en la posición 'i' y devuelve su valor."""
        if self.is_empty():
            raise IndexError("La lista está vacía")

        if i is None:
            # Si no se especifica la posición, eliminar el último elemento
            i = len(self) - 1

        if i < 0 or i >= len(self):
            raise IndexError("Posición inválida")

        if i == 0:
            value = self._head._element
            if len(self) == 1:
                # Caso especial: la lista solo tiene un elemento
                self._head = None  # La lista queda vacía
            else:
                last_node = self._head
                while last_node._next != self._head:
                    last_node = last_node._next  # Encontrar el último nodo de la lista
                self._head = self._head._next  # El siguiente de la cabeza se convierte en la nueva cabeza
                last_node._next = self._head  # El último nodo apunta a la nueva cabeza
        else:
            prev_node = self._head
            for _ in range(i - 1):
                prev_node = prev_node._next  # Encontrar el nodo anterior a la posición i
            value = prev_node._next._element  # Obtener el valor del nodo a eliminar
            prev_node._next = prev_node._next._next  # Saltar el nodo a eliminar

        self._size -= 1  # Decrementar el tamaño de la lista
        return value

    def index(self, x): 
        """Devuelve la posición de la primera aparición de 'x' en la lista."""
        if self.is_empty():
            raise ValueError("La lista está vacía")

        curr_node = self._head
        for i in range(len(self)):
            if curr_node._element == x:
                return i
            curr_node = curr_node._next

        raise ValueError("El elemento no está en la lista")
```

```python
class ListaDoblementeEnlazada:
    """Clase Lista doblemente enlazada."""

    #-------- Clase Anidada - NODO ------------#

    class _Node:
        """Clase Nodo."""
        __slots__ = '_element', '_prev', '_next' # optimiza el uso de memoria

        def __init__(self, element, prev, next): 
            self._element = element # inicializar el contenido del Nodo
            self._next = next       # referencia al siguiente Nodo
            self._prev = prev       # referencia al Nodo anterior

    def __init__(self):
        """Crea una Lista Vacia.""" 
        self._head = None  # referencia a la "cabeza" de la lista
        self._size = 0     # cantidad de elementos

    def __len__(self):
        """Retorna el numero de elementos de la Lista."""
        return self._size

    def is_empty(self):
        """Retorna VERDADERO si la Lista esta Vacia."""
        return self._size == 0

    def append(self, x):
        """Agrega un elemento al final de la lista."""
        new_node = self._Node(x, None, None)  # Crea un nuevo nodo con el elemento x
        if self.is_empty():
            self._head = new_node  # Si la lista está vacía, el nuevo nodo se convierte en la cabeza
        else:
            current = self._head
            while current._next is not None:
                current = current._next
            current._next = new_node  # Establece el enlace del último nodo al nuevo nodo
            new_node._prev = current  # Establece el enlace del nuevo nodo al último nodo
        self._size += 1  # Incrementa el tamaño de la lista

    def insert(self, i, x):
        """Agrega el elemento 'x' en la posición 'i'."""
        if i < 0 or i > self._size:
            raise IndexError('Posición inválida')  # Lanza una excepción si la posición es inválida
        new_node = self._Node(x, None, None)  # Crea un nuevo nodo con el elemento x
        if i == 0:
            if self.is_empty():
                self._head = new_node  # Si la lista está vacía, el nuevo nodo se convierte en la cabeza
            else:
                new_node._next = self._head  # Establece el enlace del nuevo nodo al actual primero
                self._head._prev = new_node  # Establece el enlace del actual primero al nuevo nodo
                self._head = new_node  # El nuevo nodo se convierte en la cabeza
        else:
            current = self._head
            for _ in range(i - 1):
                current = current._next
            new_node._prev = current  # Establece el enlace del nuevo nodo al nodo anterior
            new_node._next = current._next  # Establece el enlace del nuevo nodo al nodo siguiente
            if current._next is not None:
                current._next._prev = new_node  # Establece el enlace del nodo siguiente al nuevo nodo
            current._next = new_node  # Establece el enlace del nodo anterior al nuevo nodo
        self._size += 1  # Incrementa el tamaño de la lista

    def remove(self, x):
        """Eliminar la primera aparición de 'x' en la lista."""
        if self.is_empty():
            raise ValueError('Lista vacía')  # Lanza una excepción si la lista está vacía
        current = self._head
        while current is not None:
            if current._element == x:
                if current._prev is not None:
                    current._prev._next = current._next  # Establece el enlace del nodo anterior al siguiente nodo
                else:
                    self._head = current._next  # Si el nodo a eliminar es la cabeza, se actualiza la cabeza
                if current._next is not None:
                    current._next._prev = current._prev  # Establece el enlace del siguiente nodo al nodo anterior
                self._size -= 1  # Reduce el tamaño de la lista
                return
            current = current._next
        raise ValueError('Elemento no encontrado')  # Lanza una excepción si el elemento no se encuentra en la lista

    def pop(self, i=None):
        """Borra el elemento en la posición 'i' y devuelve su valor."""
        if i is None:
            if self.is_empty():
                raise IndexError('La lista está vacía')  # Lanza una excepción si la lista está vacía
            current = self._head
            while current._next is not None:
                current = current._next
            if current._prev is not None:
                current._prev._next = None  # Establece el enlace del nodo anterior a None
            else:
                self._head = None  # Si el nodo a eliminar es la cabeza, se actualiza la cabeza a None
            self._size -= 1  # Reduce el tamaño de la lista
            return current._element
        if i < 0 or i >= self._size:
            raise IndexError('Posición inválida')  # Lanza una excepción si la posición es inválida
        if i == 0:
            if self.is_empty():
                raise IndexError('La lista está vacía')  # Lanza una excepción si la lista está vacía
            value = self._head._element
            self._head = self._head._next  # Actualiza la cabeza al siguiente nodo
            if self._head is not None:
                self._head._prev = None  # Establece el enlace del nuevo primero al nodo anterior a None
            self._size -= 1  # Reduce el tamaño de la lista
            return value
        current = self._head
        for _ in range(i):
            current = current._next
        value = current._element
        current._prev._next = current._next  # Establece el enlace del nodo anterior al siguiente nodo
        if current._next is not None:
            current._next._prev = current._prev  # Establece el enlace del siguiente nodo al nodo anterior
        self._size -= 1  # Reduce el tamaño de la lista
        return value

    def index(self, x):
        """Devuelve la posición de la primera aparición de 'x' en la lista."""
        if self.is_empty():
            raise ValueError('Lista vacía')  # Lanza una excepción si la lista está vacía
        current = self._head
        index = 0
        while current is not None:
            if current._element == x:
                return index  # Devuelve la posición si se encuentra el elemento
            current = current._next
            index += 1
        raise ValueError('Elemento no encontrado')  # Lanza una excepción si el elemento no se encuentra en la lista
```

### Ejercicio 9

Implementar una lista *doblemente* enlazada y *circular*.

```python
class ListaDoblementeEnlazadaCircular:
    """Clase Lista doblemente enlazada circular."""

    class _Node:
        """Clase Nodo."""
        __slots__ = '_element', '_prev', '_next' # optimiza el uso de memoria

        def __init__(self, element, prev=None, next=None): 
            self._element = element  # inicializar el contenido del Nodo
            self._next = next  # referencia al siguiente Nodo
            self._prev = prev  # referencia al Nodo anterior

    def __init__(self):
        """Crea una Lista Vacia.""" 
        self._head = None  # referencia a la "cabeza" de la lista
        self._size = 0  # cantidad de elementos

    def __len__(self):
        """Retorna el numero de elementos de la Lista."""
        return self._size

    def is_empty(self):
        """Retorna VERDADERO si la Lista esta Vacia."""
        return self._size == 0

    def append(self, x):
        """Agrega un elemento al final de la lista."""
        new_node = self._Node(x)  # Crea un nuevo nodo con el elemento x

        if self.is_empty():
            new_node._next = new_node  # El nuevo nodo se enlaza a sí mismo
            new_node._prev = new_node  # El nuevo nodo se enlaza a sí mismo
            self._head = new_node  # El nuevo nodo se convierte en la cabeza
        else:
            last_node = self._head._prev  # Obtiene el último nodo
            new_node._next = self._head  # El nuevo nodo se enlaza al primero
            new_node._prev = last_node  # El nuevo nodo se enlaza al último
            self._head._prev = new_node  # El antiguo último nodo se enlaza al nuevo nodo
            last_node._next = new_node  # El antiguo último nodo se enlaza al nuevo nodo

        self._size += 1  # Incrementa el tamaño de la lista

    def insert(self, i, x):
        """Agrega el elemento 'x' en la posición 'i'."""
        if i < 0 or i > self._size:
            raise IndexError('Posición inválida')  # Lanza una excepción si la posición es inválida

        if i == self._size:  # Si se está insertando al final de la lista
            self.append(x)
            return

        new_node = self._Node(x)  # Crea un nuevo nodo con el elemento x

        if i == 0:  # Si se está insertando al principio de la lista
            new_node._next = self._head  # El nuevo nodo se enlaza al primero
            new_node._prev = self._head._prev  # El nuevo nodo se enlaza al último
            self._head._prev._next = new_node  # El antiguo último nodo se enlaza al nuevo nodo
            self._head._prev = new_node  # La cabeza se enlaza al nuevo nodo
            self._head = new_node  # El nuevo nodo se convierte en la cabeza
        else:
            current = self._head
            for _ in range(i - 1):
                current = current._next
            new_node._next = current._next  # El nuevo nodo se enlaza al siguiente nodo
            new_node._prev = current  # El nuevo nodo se enlaza al nodo actual
            current._next._prev = new_node  # El nodo siguiente se enlaza al nuevo nodo
            current._next = new_node  # El nodo actual se enlaza al nuevo nodo

        self._size += 1  # Incrementa el tamaño de la lista

    def remove(self, x):
        """Eliminar la primera aparición de 'x' en la lista."""
        if self.is_empty():
            raise ValueError('Lista vacía')  # Lanza una excepción si la lista está vacía

        current = self._head
        found = False

        while True:
            if current._element == x:
                found = True
                break
            current = current._next
            if current == self._head:
                break

        if found:
            if current._prev == current:  # Si solo hay un nodo en la lista
                self._head = None  # La lista se queda vacía
            else:
                current._prev._next = current._next  # Establece el enlace del nodo anterior al siguiente nodo
                current._next._prev = current._prev  # Establece el enlace del siguiente nodo al nodo anterior

                if current == self._head:  # Si se está eliminando la cabeza
                    self._head = current._next  # El siguiente nodo se convierte en la nueva cabeza

            self._size -= 1  # Reduce el tamaño de la lista
        else:
            raise ValueError('Elemento no encontrado')  # Lanza una excepción si el elemento no se encuentra en la lista

    def pop(self, i=None):
        """Borra el elemento en la posición 'i' y devuelve su valor."""
        if i is None:
            if self.is_empty():
                raise IndexError('La lista está vacía')  # Lanza una excepción si la lista está vacía

            value = self._head._element
            self.remove(value)
            return value

        if i < 0 or i >= self._size:
            raise IndexError('Posición inválida')  # Lanza una excepción si la posición es inválida

        if i == 0:  # Si se está eliminando la cabeza
            value = self._head._element
            self._head = self._head._next  # El siguiente nodo se convierte en la nueva cabeza

            if self._size == 1:
                self._head = None  # La lista se queda vacía
            else:
                self._head._prev = self._head._prev._prev  # El nuevo primero se enlaza al último
                self._head._prev._next = self._head  # El nuevo último se enlaza al primero

        else:
            current = self._head
            for _ in range(i):
                current = current._next

            value = current._element
            current._prev._next = current._next  # Establece el enlace del nodo anterior al siguiente nodo
            current._next._prev = current._prev  # Establece el enlace del siguiente nodo al nodo anterior

        self._size -= 1  # Reduce el tamaño de la lista
        return value

    def index(self, x):
        """Devuelve la posición de la primera aparición de 'x' en la lista."""
        if self.is_empty():
            raise ValueError('Lista vacía')  # Lanza una excepción si la lista está vacía

        current = self._head
        index = 0
        found = False

        while True:
            if current._element == x:
                found = True
                break
            current = current._next
            index += 1
            if current == self._head:
                break

        if found:
            return index
        else:
            raise ValueError('Elemento no encontrado')  # Lanza una excepción si el elemento no se encuentra en la lista

    
```
